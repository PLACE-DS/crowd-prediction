```python
def fill_nan(df_subset, column):
    df_subset.index = pd.DatetimeIndex(df_subset.index)
    
    for index, row in df_subset.iterrows():
        if index > pd.to_datetime('2020-08-31 23:45:00') and index < pd.to_datetime('2022-01-02 00:00:00'):
            if np.isnan(row[column]):
                
                seven = index - datetime.timedelta(days=7)
                fourteen = index - datetime.timedelta(days=14)
                
                seven_count = df_subset[df_subset.index == str(seven)].values[0][0]
                fourteen_count = df_subset[df_subset.index == str(fourteen)].values[0][0]
                
                if np.isnan(seven_count):
                    avg = fourteen_count
                elif np.isnan(fourteen_count):
                    avg = seven_count
                else:
                    avg = ((seven_count + fourteen_count)/2).round(0)
                row[column] = avg
            else:
                continue
    return df_subset

```

```python
df_temp_11 = fill_nan(df_small[['GAWW-11']], 'GAWW-11')
df_temp_12 = fill_nan(df_small[['GAWW-12']], 'GAWW-12')
df_temp_14 = fill_nan(df_small[['GAWW-14']], 'GAWW-14')
df_temp_17 = fill_nan(df_small[['GAWW-17']], 'GAWW-17')
```

```python
df_temp_11 = df_temp_11[(df_temp_11.index > '2020-08-31 23:45:00') & (df_temp_11.index < '2022-01-02 00:00:00')]
df_temp_12 = df_temp_12[(df_temp_12.index > '2020-08-31 23:45:00') & (df_temp_12.index < '2022-01-02 00:00:00')]
df_temp_14 = df_temp_14[(df_temp_14.index > '2020-08-31 23:45:00') & (df_temp_14.index < '2022-01-02 00:00:00')]
df_temp_17 = df_temp_17[(df_temp_17.index > '2021-04-12 23:45:00') & (df_temp_17.index < '2022-01-02 00:00:00')]
```

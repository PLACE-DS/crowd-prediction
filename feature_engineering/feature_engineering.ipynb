```python
sensors = ['GAWW-11', 'GAWW-12', 'GAWW-14']
gaww11 = cmsa.drop(['GAWW-12', 'GAWW-14'], axis=1)

gaww11.columns
```

```python
gaww11.rename(columns={'GAWW-11':'crowd_count'}, inplace=True)
gaww11.columns
```

```python
# Lagged crowd
lags = [1,2,3,4,7,12,16,20,24]
for lag in lags:
    gaww11['crowd_lag_'+str(lag)] = gaww11['crowd_count'].shift(lag)
```

```python
# Rolling mean
gaww11['rolling_mean'] = gaww11['crowd_count'].transform(lambda x: x.rolling(window=4).mean())
```

```python
# mean crowd grouped by hour
gaww11['avg_hour_crowd'] = gaww11.groupby('hour')['crowd_count'].transform('mean')

# mean crowd grouped by day of week
gaww11['avg_dow_crowd'] = gaww11.groupby('dayofweek')['crowd_count'].transform('mean')

# mean crowd grouped by month
gaww11['avg_month_crowd'] = gaww11.groupby('month')['crowd_count'].transform('mean')
```
